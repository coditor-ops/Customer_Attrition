```python
'''
It is important to create two Variables or to convert Data into 2 forms which are the
X_Data and the y_Data. X_Data contains all the features which we help our AI System to
learn and y_Data will be the Target outcome.
'''
```

```python
X_Data = Df.drop(columns='Attrition')
y_Data = Df['Attrition']

print(X_Data)
print(y_Data)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

num_cols = X_Data.select_dtypes(
    include=['int64','float64']
).columns

selector = SelectKBest(
    score_func=mutual_info_classif,
    k='all'
)

print(selector)
y = y_Data.map({
    'No':0,
    'Yes':1
})

X_num = X_Data[num_cols]

selector.fit(X_num, y)
```

```python
scores = pd.DataFrame({
    'Feature': X_num.columns,
    'Mutual_Info_Score': selector.scores_
})

scores.sort_values(
    by='Mutual_Info_Score',
    ascending=False
)
```

```python
'''
### Data Preprocessing

Before training machine learning models, the dataset must be transformed into a format that can be effectively processed by learning algorithms. This stage involves splitting the data into training and testing sets, handling different feature types, and constructing a preprocessing pipeline.

The dataset is first divided into training and testing subsets using the Train-Test Split technique. This ensures that model performance is evaluated on previously unseen data, providing a reliable estimate of its generalization ability.

Since the dataset contains both numerical and categorical features, different preprocessing techniques are applied to each type of feature. Numerical features are standardized using StandardScaler to bring all variables onto a similar scale, preventing features with larger magnitudes from dominating the learning process.

Categorical features are transformed into numerical representations using OneHotEncoder, enabling machine learning algorithms to interpret categorical information effectively.

Although no missing values were identified during exploratory data analysis, SimpleImputer is included within the preprocessing pipeline to ensure robustness and maintain compatibility with future datasets that may contain missing observations.

To streamline preprocessing and prevent data leakage, a ColumnTransformer is used to apply separate transformations to numerical and categorical features. This preprocessing workflow is then integrated into a machine learning pipeline, ensuring that all transformations are consistently applied during training, validation, and testing.
'''
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

#Converting them into 0 and 1
le = LabelEncoder()

y_Data = le.fit_transform(Df['Attrition'])

#Splitting Data into training side and the testing side
X_Train, X_Test, y_train, y_test = train_test_split(X_Data, y_Data, test_size=0.2, random_state=42, stratify=y_Data)


num_cols = X_Train.select_dtypes(
    include=['int64','float64']
).columns

cat_cols = X_Train.select_dtypes(
    include='object'
).columns

preprocessor = ColumnTransformer(
    transformers=[
        (
            'num',
            StandardScaler(),
            num_cols
        ),
        (
            'cat',
            OneHotEncoder(handle_unknown='ignore'),
            cat_cols
        )
    ]
)
X_Train_preprocessed = preprocessor.fit_transform(X_Train)
X_Test_preprocessed = preprocessor.transform(X_Test)
```

```python
'''
Models Training
Linear Models
- Logistic Regression

Distance-Based Models
- KNN
- SVM

Probabilistic Models
- Naive Bayes

Tree-Based Models
- Decision Tree
- Random Forest

Boosting Models
- Gradient Boosting
- XGBoost
'''
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier


models = {
    'Logistic Regression': LogisticRegression(),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(
        random_state=42,
    )
}

trained_models = {}

for name, model in models.items():

    model.fit(X_Train_preprocessed, y_train)

    trained_models[name] = model

    print(f"{name} Trained Successfully")


predictions = {}

for name, model in trained_models.items():

    predictions[name] = model.predict(
        X_Test_preprocessed
    )
```

```python
'''
Before Evaluating the Models on the testing data using different parameters,
we need to be make sure to check that our model is not bias or variance
Bias means that model is very easy and variance means that model has became
complex by doing the rote learning of the data.
'''
```

```python
bias_variance = []

for name, model in trained_models.items():

    train_score = model.score(
        X_Train_preprocessed,
        y_train
    )

    test_score = model.score(
        X_Test_preprocessed,
        y_test
    )

    bias_variance.append([
        name,
        round(train_score, 4),
        round(test_score, 4),
        round(train_score - test_score, 4)
    ])

bias_variance_df = pd.DataFrame(
    bias_variance,
    columns=[
        'Model',
        'Training Score',
        'Testing Score',
        'Gap'
    ]
)

bias_variance_df
```

```python
'''
The Bias-Variance analysis revealed that Decision Tree, Random Forest, and XGBoost exhibited signs of overfitting, as their training performance was significantly higher than their testing performance. This indicates high variance and reduced generalization ability on unseen data.

To further evaluate the predictive performance of each model, testing-based evaluation metrics will be employed. The models will be assessed using Accuracy Score, Precision Score, Recall Score, F1-Score, Classification Report, and Confusion Matrix.

These metrics provide a comprehensive understanding of model performance and help identify the most suitable classifier for predicting employee attrition.
'''
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

for name, model in trained_models.items():

    y_pred = model.predict(X_Test_preprocessed)

    print(f"\n{name}")
    print("-"*40)

    print(
        "Accuracy :",
        accuracy_score(y_test, y_pred)
    )

    print(
        "Precision:",
        precision_score(y_test, y_pred)
    )

    print(
        "Recall   :",
        recall_score(y_test, y_pred)
    )

    print(
        "F1 Score :",
        f1_score(y_test, y_pred)
    )
```

```python
'''
Checking that the Accuracy which we achieved from our side
is correct or not? Hence we can cross check it by cross_val
'''
```

```python
from sklearn.model_selection import cross_val_score

selected_models = {
    'Logistic Regression': LogisticRegression(),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(
        random_state=42
    )
}

for name, model in selected_models.items():

    scores = cross_val_score(
        model,
        X_Train_preprocessed,
        y_train,
        cv=5,
        scoring='accuracy'
    )

    print(f"\n{name}")
    print("-"*40)
    print("CV Scores:", scores)
    print("Mean CV Score:", scores.mean())
```
